```python
import pandas as pd
import os
import shutil
```

# Covid-19 X_ray photos

```python
FILE_PATH = 'covid-chestxray/metadata.csv'
IMAGE_PATH = 'covid-chestxray/images'
df = pd.read_csv(FILE_PATH)
df.head()
```

```python
# Get the Covid-19 X_ray with PA view
TARGET_DIR = r'Dataset\Covid'
cnt = 0
for (i, row) in df.iterrows():
    if row['finding']=='COVID-19' and row['view'] == 'PA':
        filename = row['filename']
        image_path = os.path.join(IMAGE_PATH,filename)
        image_copy_path = os.path.join(TARGET_DIR, filename)
        shutil.copy2(image_path, image_copy_path)
        cnt += 1
print(cnt)
```

# Split the data in train, test and valid

```python
# Path where the data will store
train_covid = r'Dataset\train\Covid'
test_covid = r'Dataset\test\Covid'
valid_covid = r'Dataset\valid\Covid'

train_normal = r'Dataset\train\Normal'
test_normal = r'Dataset\test\Normal'
valid_normal = r'Dataset\valid\Normal'

train_pneu = r'Dataset\train\Pneumonia'
test_pneu = r'Dataset\test\Pneumonia'
valid_pneu = r'Dataset\valid\Pneumonia'

covid = 'Dataset/Covid'
normal = 'Dataset/Normal'
pneu = 'Dataset/Pneumonia'
```

```python
# Split the covid-19 X_ray photos into train, test and valid
image_names = os.listdir(covid)
random.shuffle(image_names)
for i in range(201): #train Data
    if i <140:
        image_name = image_names[i]
        image_path = os.path.join(covid,image_name)
        target_path = os.path.join(train_covid, image_name)
        shutil.copy2(image_path, target_path)
    elif i >=140 and i<170: # test Data
        image_name = image_names[i]
        image_path = os.path.join(covid,image_name)
        target_path = os.path.join(test_covid, image_name)
        shutil.copy2(image_path, target_path)
    else: #valid Data
        image_name = image_names[i]
        image_path = os.path.join(covid,image_name)
        target_path = os.path.join(valid_covid, image_name)
        shutil.copy2(image_path, target_path)
        
```

```python
# Split the normal X_ray photos into train, test and valid
image_names = os.listdir(normal)
random.shuffle(image_names)
for i in range(500):
    if i <400: #train Data
        image_name = image_names[i]
        image_path = os.path.join(normal,image_name)
        target_path = os.path.join(train_normal, image_name)
        shutil.copy2(image_path, target_path)
    elif i>=400 and i<450: #test Data
        image_name = image_names[i]
        image_path = os.path.join(normal,image_name)
        target_path = os.path.join(test_normal, image_name)
        shutil.copy2(image_path, target_path)
    else: #valid Data
        image_name = image_names[i]
        image_path = os.path.join(normal,image_name)
        target_path = os.path.join(valid_normal, image_name)
        shutil.copy2(image_path, target_path)
```

```python
# Split the pneumoinia X_ray photos into train, test and valid
image_names = os.listdir(pneu)
random.shuffle(image_names)
for i in range(500):
    if i <400: #train Data
        image_name = image_names[i]
        image_path = os.path.join(pneu,image_name)
        target_path = os.path.join(train_pneu, image_name)
        shutil.copy2(image_path, target_path)
    elif i>=400 and i<450: #test data
        image_name = image_names[i]
        image_path = os.path.join(pneu,image_name)
        target_path = os.path.join(test_pneu, image_name)
        shutil.copy2(image_path, target_path)
    else: #valid data
        image_name = image_names[i]
        image_path = os.path.join(pneu,image_name)
        target_path = os.path.join(valid_pneu, image_name)
        shutil.copy2(image_path, target_path)
```
